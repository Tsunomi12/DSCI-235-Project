# Project Overview

The Human Development Index (HDI) is according to the United Nations, “a summary measure of average achievement in key dimensions of human development: a long and healthy life, being knowledgeable and having a decent standard of living. The HDI is the geometric mean of normalized indices for each of the three dimensions.” This system of measuring Human development Index has been around since the late 20th century and thus provides a great background to perform comparative analysis on. This is in fact a primary way that economists judge a countries standard of living rather then just relying on GDP per capita. In our project we will be performing graphical analysis on key variables that may or may not have a significant impact of HDI levels.

Importing Data Set Cleaning Data Set and importing Libraries:



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
WDVP = pd.read_csv("https://raw.githubusercontent.com/Tsunomi12/DSCI-235-Project/main/WDVP%20Datasets%202019%20-%20what%20makes%20a%20'good'%20government_.csv")
WDVP = WDVP.drop(index=[0, 1, 2, 3]) #Drop unrelated rows: source, URL, notes, data year
WDVP = WDVP.drop(['Unnamed: 4', 'Unnamed: 10', 'Unnamed: 22'], axis=1) #Dropping empty columns
print(WDVP.columns)
WDVP.head(10)
```

#### Through running this comparative analysis and relation between the many variables within our data set we will be better able to explain how certain aspects of a countries environment, social services, unemployment, gd, etc. Have a impact on our quality of living and Human development index compared between countries. 


> We will be focusing on a variety of countries that varying levels of GDP. We Will be Comparing these Countries; Canada, United States, United Kingdom, India, Ghana, Morocco and the Netherlands.





#### The First comparison we will be making is a graphial relation between Gdp Per Capita and Human Development Index between these 7 countries. 


> Higher GDP per capita means that a country has a higher income per person, which can lead to better access to basic needs such as food, housing, and healthcare. This can result in higher life expectancy, as well as better education outcomes, such as higher enrollment rates and completion rates. Additionally, higher GDP per capita can provide more resources for investment in public infrastructure, education, and healthcare, which can further improve the well-being of a country's population.



```python
WDVP.describe(exclude=['string'])
```

```python

from scipy.interpolate import make_interp_spline

# Create the figure and the two subplots
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

countries = ['United States', 'Canada', 'United Kingdom', 'India', 'Ghana', 'Morocco', 'Netherlands']
gdp_hdi_plotdata = WDVP[WDVP['indicator'].isin(countries)][['indicator','human development index', 'GDP per capita (PPP)']]
gdp_hdi_plotdata = gdp_hdi_plotdata.sort_values(by='human development index')

# Set the x-axis ticks and labels
x_ticks = np.arange(len(gdp_hdi_plotdata))
x_ticklabels = gdp_hdi_plotdata['indicator']
ax1.set_xticks(x_ticks)
ax1.set_xticklabels(x_ticklabels, rotation=45, ha='right')

# Create the spline interpolation for the HDI data
x_new = np.linspace(x_ticks.min(), x_ticks.max(), 300)
spl = make_interp_spline(x_ticks, gdp_hdi_plotdata['human development index'], k=3)
y_smooth = spl(x_new)

# Plot the lines for HDI and GDP per capita
ax1.plot(x_new, y_smooth, '-', color='green', label='HDI')
ax2.plot(x_ticks, gdp_hdi_plotdata['GDP per capita (PPP)'], 'o-', color='blue', label='GDP per capita')

# Set the y-axis labels and colors
ax1.set_ylabel('Human Development Index', color='green')
ax2.set_ylabel('GDP per capita (PPP)', color='blue')

# Set the legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper left')

# Set the title and show the plot
plt.title('Comparison of HDI and GDP per capita for selected countries')
plt.show()
```


#### Upon looking at this graph:

We can confirm our hypothesis that there is a strong correlation between GDP per capita and HDI. Upon looking at the graph we can view a strong correlation between these two variable, Ghana who had the lowest GDP per capita has the lowest HDI score and the netherlands who has the highest GDP per capita has the highest HDI.

#### The Second graphical relation we will be making is between educational expenditure per person and HDI. 



> Once of the most important essentials of the human development index is educational levels. Investing in education is essential for improving a certain countries HDI. It increasing access to health care, reduces poverty, and promotes economic growth. Higher education expenditures can contribute to the development of a skilled workforce, enhance research and development, and provide better access to education for all citizens.



```python

# Select countries and indicators of interest
gdp_educ_plotdata = WDVP[WDVP['indicator'].isin(countries)][['indicator', 'human development index', 'education expenditure\r\n% of GDP']]


# Sort data by HDI score
gdp_educ_plotdata = gdp_educ_plotdata.sort_values('human development index', ascending=True)

# Create the line chart for HDI
fig, ax = plt.subplots()
ax.plot(gdp_educ_plotdata['indicator'], gdp_educ_plotdata['human development index'], color='royalblue', marker='o', linewidth=2)


# Create the bar chart for education expenditure per person
ax2 = ax.twinx()
ax2.bar(gdp_educ_plotdata['indicator'], gdp_educ_plotdata['education expenditure\r\n% of GDP'], color='orange')


# Set the chart title and axis labels
ax.set_title('Comparison of HDI and Education Expenditure as % of GDP')
ax.set_xlabel('Country')
ax.set_ylabel('HDI')
ax2.set_ylabel('Education Expenditure per as % of GDP')


# Adjust the layout to prevent overlapping labels
fig.subplots_adjust(right=0.8)


# Display the chart
plt.show()


```


#### Summing up our graph

There is generally a positive relationship between education expenditure per person and the Human Development Index (HDI). The HDI is a composite measure of human development that takes into account a country's achievements in health, education, and income. There is a cause for more research as naturally countries with a smaller GDP will spend more % of their gdp on Education (depending on gov policies)

#### The Third Graphical relation we will be making is between unemployment rate and HDI

High unemployment rates can have negative impacts on human development outcomes such as poverty, inequality, and social exclusion. Unemployment can lead to reduced income and opportunities, which can affect access to education, healthcare, and basic needs.

```python
import seaborn as sns
gdp_unemp_plotdata = WDVP[WDVP['indicator'].isin(countries)][['indicator', 'human development index', 'unemployment (%)']]

# Sort by HDI in descending order
gdp_unemp_plotdata = gdp_unemp_plotdata.sort_values('human development index', ascending= True)

# Create the scatter plot with regression lines
sns.set_style('darkgrid')
g = sns.lmplot(x='unemployment (%)', y='human development index', data=gdp_unemp_plotdata, hue='indicator', height=7, aspect=1.2, scatter_kws={'s': 100})

# Set the chart title and axis labels
plt.title('Unemployment Rate vs. Human Development Index by Country')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Human Development Index')

# Adjust the legend position
g._legend.set_bbox_to_anchor((1, 0.5))

# Display the chart
plt.show()

```

#### Analysis of Graph:

In this graph we have an excellent explanation of an economic phenomenon called structual unemployment. Highly developed nations may experience higher levels of this structural unemployment, which is caused by a mismatch between the skills of available workers and the skills demanded by employers. Naturally there is a higher demand for skilled employers in developed nations rather than in underdeveloped nations. 

---

### The Problem with simply looking at correlation effects.

However, it is important to note that the relationship between unemployment, GDP per capita, and educational expenditure per person compared to HDI is not always straightforward. Some countries may have low unemployment rates, but still struggle with poverty and inequality due to other factors such as low wages or limited job opportunities.  Some countries with relatively high education expenditures may not perform well on the HDI due to other factors, such as income inequality or political instability.

---

### How can we rectify this?

We can do a bargraph comparison across our selected countries analyzing for each selected country, analyzing factors that are bound to have a significant impact on HDI but are commonly overlooked:

Regulatory Quality - a measure of the effectiveness, transparency, and consistency of a country's regulatory policies and institutions.

Rule of Law -  a measure that refers to the idea that all people, institutions, and entities within a society are subject to and accountable to the law, regardless of their social status

Control of Corruption -  a measure that refers to the extent to which a country's public institutions, including its government, judiciary, police, and civil service, are free from corruption and operate with integrity and transparency. 



---


On our left bar graph we will be doing a comparision of all these values where positive values are inherintely good (effective rule of law for example).

On our right bar graph we will be showing a simple graph of HDI within each country so we are able to make correlations between each country.


```python
multivar_data = WDVP[WDVP['indicator'].isin(countries)][['indicator', 'human development index', 'regulatory quality', 'rule of law', 'control of corruption']]

grouped_data1 = multivar_data.groupby('indicator')[['human development index','regulatory quality', 'rule of law', 'control of corruption']].mean()

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the barplot with three bars per country
axs[0].bar(np.arange(len(grouped_data1.index))-0.2, grouped_data1['control of corruption'], width=0.2, align='center', label='Control of Corruption')
axs[0].bar(np.arange(len(grouped_data1.index)), grouped_data1['rule of law'], width=0.2, align='center', label='Rule of Law')
axs[0].bar(np.arange(len(grouped_data1.index))+0.2, grouped_data1['regulatory quality'], width=0.2, align='center', label='Regulatory Quality')
axs[0].set_xticks(range(len(grouped_data1.index)))
axs[0].set_xticklabels(grouped_data1.index, rotation=90)
axs[0].set_ylabel('Index')
axs[0].set_title('Multi-Variable Plot per Country')
axs[0].legend(fontsize='small', loc='lower center', bbox_to_anchor=(1, -0.3))

# Plot the barplot with one bar per country
axs[1].bar(np.arange(len(grouped_data1.index)), grouped_data1['human development index'], width=0.5, align='center', label='Human Development Index')
axs[1].set_xticks(range(len(grouped_data1.index)))
axs[1].set_xticklabels(grouped_data1.index, rotation=90)
axs[1].set_ylabel('Index')
axs[1].set_title('HDI per Country')

# Show the figure
plt.show()
```

As we can see from the above graph there is a natural impact on government stabilility ( Regulatory Quality, Rule of Law, and Control of Corruption on HDI. Countries who have lower values of these variables have lower levels of HDI
